```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
df=pd.read_csv("income.csv")
df
```

```python
obj=KMeans(n_clusters=3)
y_predicted=obj.fit_predict(df[["Age","Income($)"]])
y_predicted
```

```python
df["clusters"]=y_predicted
df
```

```python
df1=df[df.clusters==0]
df2=df[df.clusters==1]
df3=df[df.clusters==2]
plt.scatter(df1["Age"],df1["Income($)"],color="red")
plt.scatter(df2["Age"],df2["Income($)"],color="yellow")
plt.scatter(df3["Age"],df3["Income($)"],color="pink")
plt.scatter(obj.cluster_centers_[:,0],obj.cluster_centers_[:,1],color="orange",label="centroid",marker="*")
```

```python
scaler=MinMaxScaler()
scaler.fit(df[["Age"]])
df["Age"]=scaler.transform(df[["Age"]])
scaler.fit(df[["Income($)"]])
df["Income($)"]=scaler.transform(df[["Income($)"]])
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[["Age","Income($)"]])
y_predicted
```

```python
df["clusters"]=y_predicted
df
```

```python
df1=df[df.clusters==0]
df2=df[df.clusters==1]
df3=df[df.clusters==2]
plt.scatter(df1["Age"],df1["Income($)"],color="red")
plt.scatter(df2["Age"],df2["Income($)"],color="green")
plt.scatter(df3["Age"],df3["Income($)"],color="blue")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],label="centroid",marker="*",color="yellow")
```

```python
sse=[]
k_range=range(1,10)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[["Age","Income($)"]])
    sse.append(km.inertia_)
plt.show()
plt.plot(k_range,sse)
```
